```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
Y_labels = ["N", "Cab", "Ccx", "Canth", "Cm", "Cw", "Cp", "CBC"]
X_labels = ['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8', 'B8A', 'B9', 'B11', 'B12']
```

First, I analyzed the relationship  between given leaf-level properties.

```python
filename = r"C:\Users\Administrator\Desktop\training data\test\processed_data.csv"
data = pd.read_csv(filename)
data
```

```python
properties = data[Y_labels]
correlation_matrix = properties.corr()
column_names = [f'Feature_{i}' for i in range(correlation_matrix.shape[1])]
# sorted_corr_matrix = correlation_matrix.abs().sort_values(by=correlation_matrix.columns.tolist(), ascending=False)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="PiYG", vmin=-1, vmax=1, square=True)
plt.title('Correlation Heatmap with Sorted Columns')
plt.show()
```

```python
spec_data = data[X_labels]
spec_data
```

```python
# Compute the correlation between the two groups
correlation_result = pd.DataFrame(index=Y_labels, columns=X_labels)
for y in Y_labels:
    for x in X_labels:
        correlation_result.loc[y, x] = properties[y].corr(spec_data[x])

# Convert the correlation result to float for proper heatmap plotting
correlation_result = correlation_result.astype(float)

# Plot the correlation heatmap between Y_labels and X_labels
plt.figure(figsize=(14, 10))
sns.heatmap(correlation_result, annot=True, cmap="PiYG", vmin=-1, vmax=1)
plt.title('Correlation Heatmap Between Group 1 and Group 2')
plt.show()
```

```python
# Correlation Pairplot for Original Data
sns.pairplot(properties, diag_kind='kde', kind='scatter')
plt.suptitle('Pairplot of Original Data', y=1.02)
plt.show()
```
